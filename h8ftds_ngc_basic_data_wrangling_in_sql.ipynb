```python
from google.colab import auth
from google.cloud import bigquery
import pandas as pd

auth.authenticate_user()
project_id = "excellent-camp-382622"
client = bigquery.Client(project=project_id)
```

How many tables are in the austin_bikeshare dataset?

```python
query = """SELECT COUNT(*) AS total_tables FROM  `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.TABLES`"""
query_job = client.query(query).to_dataframe()
query_job
```

How many columns in the bikeshare_stations table have STRING data?

```python
query = """SELECT COUNT(*) AS total_columns_str_type FROM `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS` WHERE table_name = 'bikeshare_stations' AND data_type = 'STRING'"""
query_job = client.query(query).to_dataframe()

query_job
```

Retrieve all trips where the bike_id is not null and the duration is greater than 1000 minutes

```python
query = """SELECT * FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` WHERE bike_id IS NOT NULL AND duration_minutes > 1000"""
query_job = client.query(query).to_dataframe()

query_job
```

Filter trips where the suscriber_type is "Walk Up" and the end_station_name  is "Stolen"

```python
query = """SELECT * FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` WHERE subscriber_type = 'Walk Up' AND end_station_name = 'Stolen'"""
query_job = client.query(query).to_dataframe()

query_job
```
