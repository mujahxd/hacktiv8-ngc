```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
# Ganti 'path_to_chromedriver' dengan jalur lengkap ke chromedriver di komputer Anda
driver = webdriver.Chrome()
driver.maximize_window()

name, price, market_cap, volume, coin_supply = [], [], [], [], []


for i in range(1, 11):
    url = 'https://coinmarketcap.com/?page={}'.format(i)
    driver.get(url)

# Tunggu hingga elemen yang diinginkan dimuat
    time.sleep(15)  # Ubah sesuai kebutuhan

# Simulasikan pengguliran halaman beberapa kali
    scroll_iterations = 16  # Ubah sesuai kebutuhan
    for _ in range(scroll_iterations):
    # Gulir ke bawah halaman
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
#     driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_DOWN)
    
    # Tunggu sebentar antara pengguliran
        time.sleep(2)  # Ubah sesuai kebutuhan

# Ambil sumber HTML halaman menggunakan requests atau httpx
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    for tr in soup.tbody:
        if tr.find('p', class_='sc-4984dd93-0 kKpPOn'):
            name.append(tr.find('p', class_='sc-4984dd93-0 kKpPOn').get_text())
        else:
            name.append(pd.NA)
        
        if tr.find('span', class_='sc-f8982b1f-1 bOsKfy'):
            market_cap.append(tr.find('span', class_='sc-f8982b1f-1 bOsKfy').get_text())
        else:
            market_cap.append(pd.NA)
            
        if tr.find('div', class_='sc-a0353bbc-0 gDrtaY'):
            price.append(tr.find('div', class_='sc-a0353bbc-0 gDrtaY').get_text())
        else:
            price.append(pd.NA)
        
        if tr.find('p', class_='sc-4984dd93-0 jZrMxO font_weight_500'):
            volume.append(tr.find('p', class_='sc-4984dd93-0 jZrMxO font_weight_500').get_text())
        else:
            volume.append(pd.NA)
        
        if tr.find('p', class_='sc-4984dd93-0 WfVLk'):
            coin_supply.append(tr.find('p', class_='sc-4984dd93-0 WfVLk').get_text())
        else:
            coin_supply.append(pd.NA)
    
driver.quit()
    
data_multiplepage = pd.DataFrame()
data_multiplepage['name'] = name
data_multiplepage['price'] = price
data_multiplepage['market_cap'] = market_cap
data_multiplepage['volume'] = volume
data_multiplepage['coin_supply'] = coin_supply
```

```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By


driver = webdriver.Chrome()
driver.maximize_window()

name = []
temperature = []
feels_like = [] 
wind_speed = [] 
wind_direction = [] 
humidity = [] 
dew_point = [] 
visibility = [] 
probability_of_rain = [] 
amount_of_rain = []
    
url = 'https://www.timeanddate.com/weather/indonesia'
driver.get(url)

time.sleep(10)  

scroll_iterations = 2  
for _ in range(scroll_iterations):
   
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
    time.sleep(2)  


html = driver.page_source


soup = BeautifulSoup(html, 'html.parser')
    
table_weather = soup.find('table', class_='zebra fw tb-wt zebra va-m')

for tag in table_weather.tbody.find_all('a'):
    name.append(tag.get_text())
    link = "https://www.timeanddate.com" + tag['href']
    
#     driver.maximize_window()
    driver.get(link)
    time.sleep(10)
    
    scroll_iterations = 2  
    for _ in range(scroll_iterations):
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
        time.sleep(2)
        
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    
    rows_of_table = soup.find('table', class_='zebra tb-wt tc sep').tbody.find_all('tr')

    if rows_of_table[1].find('td'):
        temperature.append(rows_of_table[1].find('td').get_text().replace('\xa0', ' ')) #temperature
    else:
        temperature.append(pd.NA)
        
    if rows_of_table[3].find('td'):
        feels_like.append(rows_of_table[3].find('td').get_text().replace('\xa0', ' ')) #feels like
    else:
        feels_like.append(pd.NA)
    
    if rows_of_table[4].find('td'):
        wind_speed.append(rows_of_table[4].find('td').get_text()) # wind speed
    else:
        wind_speed.append(pd.NA)
    
    if rows_of_table[5].find('td'):
        wind_direction.append(rows_of_table[5].find('td').get_text()[0]) # wind direction
    else:
        wind_direction.append(pd.NA)
    
    if rows_of_table[6].find('td'):
        humidity.append(rows_of_table[6].find('td').get_text()) #humidity
    else:
        humidity.append(pd.NA)
    
    if rows_of_table[7].find('td'):
        dew_point.append(rows_of_table[7].find('td').get_text().replace('\xa0', ' ')) # dew point
    else:
        dew_point.append(pd.NA)
    
    if rows_of_table[8].find('td'):
        visibility.append(rows_of_table[8].find('td').get_text().replace('\xa0', ' ')) # visibility
    else:
        visibility.append(pd.NA)
    
    if rows_of_table[9].find('td'):
        probability_of_rain.append(rows_of_table[9].find('td').get_text()) # probability
    else:
        probability_of_rain.append(pd.NA)
    
    # amount of rain
    try:
        if rows_of_table[10].find('td'):
            amount_of_rain.append(rows_of_table[10].find('td').get_text())
    except IndexError:
        amount_of_rain.append(pd.NA)
        
data_cuaca = pd.DataFrame()
data_cuaca['name'] = name
data_cuaca['temperature'] = temperature
data_cuaca['feels_like'] = feels_like
data_cuaca['wind_speed'] = wind_speed
data_cuaca['wind_direction'] = wind_direction
data_cuaca['humidity'] = humidity
data_cuaca['dew_point'] = dew_point
data_cuaca['visibility'] = visibility
data_cuaca['probability_of_rain'] = probability_of_rain
data_cuaca['amount_of_rain'] = amount_of_rain
```
